```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Laginho/SmartCity-Simulator/refs/heads/main/src/Input/dados_onibus.csv")
```

```python
df_linha1 = df[df['linha'] == 1]
df_linha1.index = df_linha1["dia"]

df_linha2 = df[df['linha'] == 2]
df_linha2.index = df_linha2["dia"]

df_linha3 = df[df['linha'] == 3]
df_linha3.index = df_linha3["dia"]

df_linha4 = df[df['linha'] == 4]
df_linha4.index = df_linha4["dia"]

df_linha5 = df[df['linha'] == 5]
df_linha5.index = df_linha5["dia"]
```

```python
df_linha1['requests'].plot(title='Requisições das linhas')
df_linha2['requests'].plot(linestyle = "--", marker = "o")
df_linha3['requests'].plot(linestyle = ":", marker = "o")
plt.legend(['Linha 1', 'Linha 2', 'Linha 3'])
```

```python
df_linha3['requests'].plot(title='Requisições das linhas', color = "green")
df_linha4['requests'].plot(linestyle = "-.", color = "blue", marker = "o")
df_linha5['requests'].plot(linestyle = "--", color = "black", marker = "o")
plt.legend(['Linha 3', 'Linha 4', 'Linha 5'])
```
